## Load BART Model and Tokenizer

### Subtask:
Load the `facebook/bart-large-cnn` pre-trained model and its corresponding tokenizer from the `transformers` library. This model is fine-tuned for text summarization.


**Reasoning**:
To load the BART model and tokenizer, I need to use the `pipeline` function from the `transformers` library and specify the model name `facebook/bart-large-cnn` for the 'summarization' task.



```python
from transformers import pipeline

# Initialize a summarization pipeline with the BART large CNN model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

print("BART summarization pipeline loaded successfully.")
```

**Reasoning**:
Now that the summarization pipeline is loaded, the next logical step is to provide an example text and use the `summarizer` to generate a summary.



```python
example_text = (
    "The Orbiter Discovery, commanded by veteran astronaut Charlie Bolden, has completed its mission to service the Hubble Space Telescope. The seven-member crew performed five spacewalks over 11 days to install new instruments and repair existing ones, significantly extending the observatory's lifespan and improving its capabilities. This mission, the fifth and final servicing mission to Hubble, was crucial for maintaining the telescope's status as a premier tool for astronomical research. Among the upgrades were a new Wide Field Camera 3, designed to capture stunning, wide-angle views of the cosmos, and the Cosmic Origins Spectrograph, which will allow astronomers to study the invisible light from distant objects. The successful repairs and upgrades are expected to enable Hubble to continue its groundbreaking discoveries for at least another five to ten years."
)

# Generate summary
summary_results = summarizer(example_text, max_length=150, min_length=30, do_sample=False)

# Print the summary
print("Original Text:\n", example_text)
print("\nSummarized Text:\n", summary_results[0]['summary_text'])
```

## Run Streamlit App

Run the created Streamlit application in the Colab environment and get the public URL to access it.

```python
%%writefile app.py
import streamlit as st
from transformers import pipeline

# Load the summarization pipeline (cached to avoid re-loading on each rerun)
@st.cache_resource
def load_summarizer():
    return pipeline("summarization", model="facebook/bart-large-cnn")

summarizer = load_summarizer()

st.title("Text Summarization with BART-large-CNN")
st.markdown("Enter text below to get a summarized version using the `facebook/bart-large-cnn` model.")

# Text area for user input
input_text = st.text_area("Enter text here:", height=250)

# Slider for summary length
min_length = st.slider("Minimum Summary Length", min_value=10, max_value=100, value=30)
max_length = st.slider("Maximum Summary Length", min_value=min_length + 10, max_value=500, value=150)

if st.button("Summarize"):
    if input_text:
        with st.spinner("Generating summary..."):
            try:
                summary_results = summarizer(
                    input_text,
                    max_length=max_length,
                    min_length=min_length,
                    do_sample=False
                )
                st.subheader("Generated Summary:")
                st.write(summary_results[0]['summary_text'])
            except Exception as e:
                st.error(f"An error occurred during summarization: {e}")
    else:
        st.warning("Please enter some text to summarize.")
```

```python
from pyngrok import ngrok

# Kill any existing ngrok processes
ngrok.kill()

# Set your Ngrok authtoken (replace 'your_auth_token' with the copied token)
ngrok.set_auth_token('')

# Set up the Ngrok tunnel to the Streamlit app
# The port number should be included in the 'addr' argument
public_url = ngrok.connect(addr='http://localhost:8501')
print(f"Streamlit app is live at: {public_url}")
```
